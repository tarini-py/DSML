## <font color='blue'>**Basic Terminologies**</font>



### <font color='purple'>**1. Experiment**</font>
- It is basically an activity which I'm trying to do.  

> **Let's say I have this mathematical  equation**  

  $a^2 + b^2 + 2ab$    

where: $a$ = 3 and $b$ = 4    

$3^2 + 4^2 + 2(3)(4)$ = 49



- We are 100% sure that the result of this equation will be 49 only. It cannot be 50 or 48.

This type of experiment is called **Deterministic Experiments** where we can **determine** the exact output, like in this case.  

<br>  

Now, let's see another few more examples:

- <font color='orange'>Flipping a coin  </font>
 - When you flip a coin, there are two possible outcomes: it can land either `heads` or `tails`.   

- <font color='orange'>Rolling a six-sided die  </font>
 - When you roll the die, the outcome is uncertain, and the die can land on any of the six faces.  

- <font color='orange'>Cricket Match  </font>
 - Suppose there is a match going on between 2 teams, we can't determine the match result.

<br>

In all of these above examples, we can notice one common thing.   

> <font color='purple'>**Q. Can we determine the outcome of all these experiments?**</font>

No, because the outcomes are uncertain. These types of experiments are known as **Probabilistic Experiments**.





Let's continue with the experiment of "Rolling a six sided die" and look at the possible results of an experiment.  

#### <font color='orange'>**Experiment: Rolling a die**</font>



### <font color='purple'>**2. Outcomes**</font>

  - Suppose we roll a six sided die and we want to know the possible `Outcomes`.

  - We know that we could get any digit out of the 6 digits. So, an outcome could be : {1} or {2} or {3} or {4} or {5} or {6}  



### <font color='purple'>**3. Sample Space**</font>  
   
- It is the collection of all the possible outcomes of the experiment.   

So the **sample space** for this experiment will be: **{1, 2, 3, 4, 5, 6}**  


### <font color='purple'>**4. Events**</font>

We know that sample space for die is {1,2,3,4,5,6}.

If we say,

> <font color='purple'>**An Even number is rolled / While rolling a die, an even number has occurred**</font>  

- Then the possible outcomes will be: `{2, 4, 6}`  
<Br>

This is known as an **`Event`**.   
Any subset of sample space is an event.  
- {2, 4, 6} is a subset of sample space.   


"**An Even number is rolled**" is an event here and its output is $E$ = {2, 4, 6}, where $E$ denotes an Event.

<br>  



> <font color='purple'>**Q1. What are the possible outcomes when a dice is rolled and a number greater than two has occurred?** </font>

- For this Event, outcome will be $E$ = {3, 4, 5, 6}

Let's recall the experiment "**Rolling a die**" for which the **Sample space** is $\left \{1, 2, 3, 4, 5, 6\right\}$

### <font color='purple'>**Intersection** </font>

> <font color='purple'>**Q. In which condition, both Mohit and Rakesh will win their bets?**</font>

We want a number which occurs in both of their events

They will win their bets when we get a number 1 or 5 on a die.

- Therefore $\left\{1, 5\right\}$ is the possible outcome such that both Mohit and Rakesh will win their bets  

This is known as an <font color='purple'>**Intersection**</font> of two events.
- It is denoted as $A âˆ© B$  

- Intersection means **members belonging to both A `AND` B**  

  -  So, $A âˆ© B$ will consists only of the elements present in both events, which in this case are $\left\{1, 5\right\}$

### <font color='purple'>**Union**  </font>

Now the next question,

> <font color='purple'>**Q. When either Mohit or Rakesh will win their bets?**</font>

If we get any number out of 1, 3, 5 or 6  

- Possible outcomes of this event: $\left\{1, 3, 5, 6\right\}$   

<br>

This is known as <font color='purple'>**Union**</font> of Two events A and B  

- It is denoted by **$A âˆª B$**
- So, **Union** means **members belonging to either A `OR` B**  
- So, $A âˆª B$ will combine their outcomes, which in this case will be  $\left\{1, 3, 5, 6\right\}$

Mohit will lose his bet if the outcome is $\left\{2, 4, 6\right\}$  

This is known as <font color='purple'>**complement** </font>of Event A, denoted by $A'$ or $A^c$

We can define it as the set that contains all the elements except the elements of A, denoted as **$A' = U - A$**   


<br>

While Rakesh will lose if the outcome is $\left\{2, 3, 4\right\}$
- Hence $B' = \left\{2, 3, 4\right\}$



## <font color='blue'>**Mutually Exclusive Events (Disjoint Events)**</font>

> <font color='purple'>**Q1. What will be the output of $A âˆ© C$ ?**  </font>

We will have an empty set $\left\{  \ \right\}$  which can also be represented by $âˆ…$

Because there are no common elements in Set $A$ and Set $C$  

Or it implies that <font color='purple'>**both the events can't occur on the same time**</font> means we can't get an **Even number and a Odd number** at the same time on the dice.
  
  
- So, when two events cannot occur at the same time or simultaneously then these types of events are known as <font color='purple'>**`Mutually Exclusive Events`**</font> or **`Disjoint Events`**

## <font color='blue'>**Exhaustive Events**  </font>

> <font color='purple'>**Q. What will be the output of $AâˆªBâˆªC$?**  </font>

Our events are:  

  - $A=\left\{1,3,5\right\}$, $B=\left\{1, 5, 6\right\}$, $C=\left\{2,4,6\right\}$  

    - Therfore $AâˆªBâˆªC$ = combined elements of Event A, B, C  = {1, 2, 3, 4, 5, 6}

<br>

This is nothing but the **Sample Space** of our experiment **"Rolling a die"** as these events when combined, giving the all possible outcomes.

- These types of events are known as <font color='purple'>**`Exhaustive Events`**  </font>



## <font color='blue'>**Non Mutually Exclusive Events (Joint Events)** </font>

Suppose we define one more Events:  

- **Event D**: Rolling a number greater than 3 = (4, 5, or 6).  

<br>

> <font color='purple'>**Q. Can we say that Events C (getting even) and D are mutually exclusive?**  </font>

No, as we can get a number that is <font color='purple'>both even and greater than 3</font>,  which means both **events C and D can occur simultaneously**.   


- For instance, if the die shows a 4 or a 6, it fulfills the criteria for both events C and D.   
- This type of events are known as <font color='purple'>**`non-mutually exclusive`** </font> or **`joint events`**

non mutually exclusive.PNG

## <font color='blue'>**Independent Events**  </font>

While non-mutually exclusive events allow for overlap, where more than one event can occur, independent events focus on how the occurrence of one event **may or may not affect** the likelihood or outcome of another event  





**Suppose we have 2 two events:**  

- **Event A**: Rolling an even number (2, 4, or 6)  

- **Event B**: Flipping a coin and getting heads

<br>

> <font color='purple'>**Q. Are these two events Independent or not?** </font>

YES, these events are **`independent Events`** because   

  - The outcome of rolling the die (**Event A**) `does not affect the outcome` of flipping the coin (**Event B**), and vice versa.
  
They are unrelated events that are occurring independently.    

And if two events A and B are independent, then the probability of happening of both A and B is:  

- **$P(Aâˆ©B) = P(A) * P(B)$**  

<br>  

In case of Disjoint events, **$P(Aâˆ©B)$ = 0** , as <font color='purple'>A Intersect B = { }</font>  

- **<font color='purple'>So, if the Events are Independent they cannot be Mutually Exclusive or Disjoint and vice a versa</font>**

## <font color='blue'>**How to calculate Probability**</font>

Now if I want to calculate the Probability of the particular event let's say event A,  then we can calculate using this.  

**$Probability = \Large\frac{Outcomes  \ in \ set \ A}{Total \ Outcomes \ in \ Entire \ Sample \ Space}$**   

Now, let's take a <font color='orange'>random Experiment whose **outcome**</font> could be **{1} or {2} or {3} or {4} or {5} or {6}**, then the **Sample Space** will be **{1, 2, 3, 4, 5, 6}**    

<br>  

Let's define some events:   

1. **$A$ = {2, 4, 6}**  
  > <font color='purple'>**Q1. What will be the probability of Event A?**  </font>

  - By looking into the formula = $\Large\frac{Possible \ oucomes}{Total \ outcomes}$    
  - Possible outcomes of event A  = 3 and total Outcome in sample space = 6
  
  So, $P(A) = $$\Large\frac{3}{6}$  

2. **$B$ = {1, 2}**    
  - Similarly <font color='purple'>Probability of Event B will be</font> $P(B)$ = $\Large\frac{2}{6}$
  
3. **$C$ = {1, 4, 5, 6}**    

  - and <font color='purple'>Probability of Event C will be</font> $P(C)$ = $\Large\frac{4}{6}$



## <font color='blue'>**Addition Rule**</font>

> <font color='purple'>**Q1. What will be the Probability of $P(AâˆªB)$?**  </font>

First we need to find $A \cup B$ which is **{1, 2, 4, 6}**  

- So by the formula of probability $P(AâˆªB)$ will be = $\Large\frac{|AâˆªB|}{|S|}$ = $\Large\frac{ |\{1, 2, 4, 6\}|}{|\{1, 2, 3, 4, 5, 6\}|}$ = $\Large\frac{4}{6}$

Where, |$AâˆªB$| = Number of elements(cardinality) of ($AâˆªB$) set, <br>and |$S$| = Number of elements in Sample Space

<br>





If we want to represent using venn Diagram:  

union venn.PNG

> <font color='purple'>**Q2. What will be Probability of $P(Aâˆ©B)$?**  </font>

$Aâˆ©B$ will be **{2}**  

- So by the formula of probability $P(Aâˆ©B)$ will be = $\Large\frac{ |\{2\}|}{|\{1, 2, 3, 4, 5, 6\}|}$ = $\Large\frac{1}{6}$  

So by looking into Venn diagram, we observe that AUB means <font color='orange'>addition of all the elements of **Set A** and **Set B** </font>


- We can also notice in set A we have {2, 4, 6} and in set B we have {1, 2}    


- While adding the outcomes of the sets, {2} is occuring twice, which is nothing but $Aâˆ©B$, so we have to subtract it once from our addition, as we want unique outcomes only (Since a set can only have distinct elements).

<br>

So the formula for $P(A âˆª B)$ can we written as:

  - **$P(AâˆªB) = P(A) + P(B) - P(Aâˆ©B)$**  

This is known as **`Addition Rule`**.
This is for Joint Events  

<br>  

In case of **Disjoint Events**

  - the intersection of $A âˆ©B = \{ \ \}$ so, $P(Aâˆ©B) = 0$  

    - therefore, **$P(AâˆªB) = P(A) + P(B)$**

### <font color='purple'>**Cross Tab and contingency table**</font>

> <font color='purple'> **Q. Do you remember pivot table from DAV-1 Libraries module?** </font>


- There is a function called `pd.crosstab()` , which accepts parameters **index** and **columns**.  



```python
pd.crosstab(index=df_sachin["century"],
            columns=df_sachin["Won"],
            margins=True)
```

What we did using `.valuecounts()` at above, `pd.crosstab()` did the same thing but converted the output into nice tabular format

- **Century** is taken as the **index** and **Won** is taken as **columns**    

-   When we do **Margins = True** we get **All**, both in rows and columns,

  - The values of **All** in a ROW represnts the **Total Value** of each columns (False, True, All)  

  - The values of **All** in a COLUMN represents the **Total Value** of each rows (False, True, All)

This table is also known as <font color='purple'> **`Contingency Table`** </font>

We can calculate probabilities using the contingency table.
