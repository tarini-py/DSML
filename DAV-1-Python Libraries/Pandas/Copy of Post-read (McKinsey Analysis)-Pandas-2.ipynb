```python
import numpy as np
import pandas as pd

df=pd.read_csv("mckinsey.csv")
df
```

```python
df.describe()
```

**Conditional Filtering**: To check if there is a trend where higher GDP per capita is associated with higher life expectancy.

```python
high_gdp = df[df['gdp_cap'] > df['gdp_cap'].median()]
low_gdp = df[df['gdp_cap'] <= df['gdp_cap'].median()]

high_gdp_life_expectancy = high_gdp['life_exp'].describe()
low_gdp_life_expectancy = low_gdp['life_exp'].describe()

print(high_gdp_life_expectancy)
```

```python
print(low_gdp_life_expectancy)
```

**Insights**:
- Countries with higher GDP per capita have a higher average life expectancy compared to those with lower GDP per capita.

This analysis helps to understand the relationship between GDP per capita and life expectancy, suggesting that higher GDP is generally associated with higher life expectancy.

**Ratio Analysis:** The ratio can give you insights into how life expectancy changes relative to GDP per capita.

```python
df['Life Expectancy per GDP'] = df['life_exp'] / df['gdp_cap']
life_gdp_ratio_summary = df['Life Expectancy per GDP'].describe()

print(life_gdp_ratio_summary)
```

**Insights**:

- Higher Ratio: Countries with a higher Life Expectancy per GDP ratio achieve higher life expectancy relative to their GDP per capita. This could indicate that these countries are more efficient in translating GDP into better health outcomes.
- Lower Ratio: Countries with a lower ratio may have lower life expectancy for each unit of GDP, suggesting that additional GDP may not be as effective in improving life expectancy in these countries.
