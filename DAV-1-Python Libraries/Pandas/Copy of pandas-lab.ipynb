```python
def solve(director_name,start_year,end_year):
  movies_name = []
  movies = data[(data['director_name']==director_name) & (data['year']>=start_year) & (data['year']<=end_year)]['title']
  movies_name.append((director_name,movies))
  return movies_name
```

```python
solve('Gore Verbinski',2005,2010)
```

```python
solve('komal sharma',2005,2010) #deal with empty case => "No director / no movies detected"
```
