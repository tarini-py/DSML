```python
window_size = 3
window_sums = a[:len(a) - window_size + 1] + a[1:len(a) - window_size + 2] + a[2:len(a) - window_size + 3]

# Get max sum and the starting index
max_sum = np.max(window_sums)
start_index = np.argmax(window_sums)

# Extract the 3 elements that form the max sum
max_triplet = a[start_index:start_index + window_size]

print("Max 3-consecutive sum:", max_sum)
print("Triplet:", max_triplet)
```
