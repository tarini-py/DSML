```python
x_val = data['Genre'].value_counts().index
y_val = data['Genre'].value_counts().values
```

```python
plt.plot(x_val,y_val) #line chart
```

```python
#bar plot
plt.bar(x_val,y_val)
plt.title('genre vs count')
plt.xlabel('genre')
plt.ylabel('counts')
plt.show()
```

```python
#how to provide more customization
plt.figure(figsize = (15,10)) #(x,y)
plt.bar(x_val,y_val)
plt.title('genre vs count')
plt.xlabel('genre')
plt.ylabel('counts')
plt.show()
```

```python
plt.figure(figsize = (10,5)) #(x,y)
plt.bar(x_val,y_val,width=0.2,color='green')
plt.title('genre vs count')
plt.xlabel('genre')
plt.ylabel('counts')
plt.xticks(rotation = 70)
plt.yticks(rotation = 180)
plt.show()
```

```python
#seaborn
plt.figure(figsize=(8,5))
sns.countplot(data = data,x = 'Genre') #barplot
plt.xticks(rotation = 90)
plt.show()
```
